```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

Plain Clustering: Bar Graph

```python
AVG = {'k_means':0.255, 'agglom':0.257, 'dbscan':0.17, 'mean_shift':0.17, 'birch':0.17, 'affinity':0.394, 'mini_b_k':0.16}
algorithms = list(AVG.keys())
avg_metrics = list(AVG.values())

fig = plt.figure(figsize = (6.5, 2))

plt.bar(algorithms, avg_metrics, color ='maroon', width = 0.5)

plt.xlabel("Clustering Algorithms")
plt.ylabel("Avg. evaluation metric")
plt.title("Performance in Plain Clustering")
plt.show()
```
